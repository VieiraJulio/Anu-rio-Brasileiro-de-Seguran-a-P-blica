```python
id = 'learned-shell-439223-i1'

query = """
  SELECT
    dados.quantidade_mortes_intervencao_policial_civil_fora_de_servico as quantidade_mortes_intervencao_policial_civil_fora_de_servico,
    dados.quantidade_feminicidio as quantidade_feminicidio,
    dados.quantidade_mortes_intervencao_policial_militar_fora_de_servico as quantidade_mortes_intervencao_policial_militar_fora_de_servico,
    dados.quantidade_furto_veiculos as quantidade_furto_veiculos,
    dados.quantidade_mortes_intervencao_policial_civil_em_servico as quantidade_mortes_intervencao_policial_civil_em_servico,
    dados.quantidade_estupro as quantidade_estupro,
    dados.quantidade_morte_policiais_civis_confronto_em_servico as quantidade_morte_policiais_civis_confronto_em_servico,
    dados.quantidade_mortes_intervencao_policial_militar_em_servico as quantidade_mortes_intervencao_policial_militar_em_servico,
    dados.quantidade_mortes_policiais_confronto as quantidade_mortes_policiais_confronto,
    dados.quantidade_mortes_violentas_intencionais as quantidade_mortes_violentas_intencionais,
    dados.quantidade_posse_uso_entorpecente as quantidade_posse_uso_entorpecente,
    dados.quantidade_morte_policiais_militares_fora_de_servico as quantidade_morte_policiais_militares_fora_de_servico,
    dados.quantidade_morte_policiais_civis_fora_de_servico as quantidade_morte_policiais_civis_fora_de_servico,
    dados.quantidade_latrocinio as quantidade_latrocinio,
    dados.quantidade_porte_ilegal_arma_de_fogo as quantidade_porte_ilegal_arma_de_fogo,
    dados.quantidade_mortes_intervencao_policial as quantidade_mortes_intervencao_policial,
    dados.ano as ano,
    dados.quantidade_roubo_furto_veiculos as quantidade_roubo_furto_veiculos,
    dados.quantidade_posse_ilegal_arma_de_fogo as quantidade_posse_ilegal_arma_de_fogo,
    dados.quantidade_lesao_corporal_dolosa_violencia_domestica as quantidade_lesao_corporal_dolosa_violencia_domestica,
    dados.quantidade_trafico_entorpecente as quantidade_trafico_entorpecente,
    dados.quantidade_roubo_veiculos as quantidade_roubo_veiculos,
    dados.quantidade_lesao_corporal_morte as quantidade_lesao_corporal_morte,
    dados.proporcao_mortes_intenvencao_policial_x_mortes_violentas_intencionais as proporcao_mortes_intenvencao_policial_x_mortes_violentas_intencionais,
    dados.quantidade_morte_policiais_militares_confronto_em_servico as quantidade_morte_policiais_militares_confronto_em_servico,
    dados.quantidade_posse_ilegal_porte_ilegal_arma_de_fogo as quantidade_posse_ilegal_porte_ilegal_arma_de_fogo,
    dados.sigla_uf AS sigla_uf,
    diretorio_sigla_uf.nome AS sigla_uf_nome,
    dados.id_municipio AS id_municipio,
    diretorio_id_municipio.nome AS id_municipio_nome,
    dados.grupo as grupo,
    dados.quantidade_homicidio_doloso as quantidade_homicidio_doloso
FROM `basedosdados.br_fbsp_absp.municipio` AS dados
LEFT JOIN (SELECT DISTINCT sigla,nome  FROM `basedosdados.br_bd_diretorios_brasil.uf`) AS diretorio_sigla_uf
    ON dados.sigla_uf = diretorio_sigla_uf.sigla
LEFT JOIN (SELECT DISTINCT id_municipio,nome  FROM `basedosdados.br_bd_diretorios_brasil.municipio`) AS diretorio_id_municipio
    ON dados.id_municipio = diretorio_id_municipio.id_municipio
"""
```

```python
# pip install pandas-gbq
import pandas as pd
import pandas_gbq as pd_gbd

from cycler import cycler
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')


cores = plt.get_cmap("Paired").colors
ciclo_cores = cycler("color", cores)
plt.rc("axes", prop_cycle = ciclo_cores)
```

```python
df = pd_gbd.read_gbq(query, project_id= id , dialect='standard')
```

```python
# ETL 
colunas_excluidas = [
       'quantidade_furto_veiculos',
       'quantidade_estupro',
       'quantidade_posse_uso_entorpecente',
        'quantidade_porte_ilegal_arma_de_fogo',
       'quantidade_roubo_furto_veiculos',
       'quantidade_posse_ilegal_arma_de_fogo',
       'quantidade_lesao_corporal_morte',
       'proporcao_mortes_intenvencao_policial_x_mortes_violentas_intencionais',
       'quantidade_posse_ilegal_porte_ilegal_arma_de_fogo', 'sigla_uf',
        'id_municipio',
        'id_municipio_nome',
        'quantidade_roubo_veiculos',
        'quantidade_lesao_corporal_dolosa_violencia_domestica',
        'quantidade_trafico_entorpecente'
        ]

df = df.drop(columns= colunas_excluidas, axis = 1)
```

```python
df['ano'] = df['ano'].astype('object')
```

```python
colunas_numericas = list(df.select_dtypes('number'))

colunas_numericas
```

```python
df['Mortes totais'] = df[colunas_numericas].sum(axis = 1)
```

```python
df_final = df.drop(columns = colunas_numericas)

df_final.rename(columns= {'sigla_uf_nome' : 'Estado'}, inplace= True)
```

```python
df_final
```

```python
df_final.groupby(['Estado'])['Mortes totais'].sum().sort_values(ascending=False).to_frame()
```

```python
df_final.groupby(['ano', 'Estado'])['Mortes totais'].sum().sort_values(ascending=False).to_frame()
```

```python
df_final
```

```python
df_final = df_final.sort_values(by='Mortes totais', ascending=False)

df_final
```

```python
df_agrupado = df_final.groupby('Estado')['Mortes totais'].sum().sort_values(ascending=False).to_frame()

df_agrupado = df_agrupado.head()

dados_ordenados = sorted(zip(df_agrupado['Mortes totais'], df_agrupado.index), reverse = True)
mortes_ordenadas = [x[0] for x in dados_ordenados]
estados_ordenados = [x[1] for x in dados_ordenados]


df_agrupado
```

```python
mosaico = 'AA;BC;DD'

espacamento = {'wspace': 0.5, 'hspace': 0.5}

fig = plt.figure(figsize = (12,14))
axs = fig.subplot_mosaic(mosaico,gridspec_kw= espacamento)


# ----------------------------------------

#gráfico 1 - 

df_agrupado_ano = df_final.groupby('ano')['Mortes totais'].sum().to_frame()

#df_agrupado_ano['Mortes totais'] = df_agrupado_ano['Mortes totais'].map('{:,}'.format)

barras = axs['A'].plot(df_agrupado_ano.index, 'Mortes totais', data = df_agrupado_ano, color = 'black', marker = 'o')

rotulo = df_agrupado_ano['Mortes totais'].map('{:,}'.format)

for posicao in range(len(df_agrupado_ano.index)):
    axs['A'].annotate(rotulo.iloc[posicao], 
                   color = 'black', xy=(df_agrupado_ano.index[posicao], df_agrupado_ano.iloc[posicao]), xytext=(-5,15), textcoords= 'offset points', fontweight = 'bold', fontsize = 12)


axs['A'].text(2016, 25000,'     Pico de Violência', color = cores[5], fontweight = 'bold', fontsize = 9)
axs['A'].axvspan(2016,2017, alpha = 0.3, color = cores[4])

axs['A'].set_xticks(df_agrupado_ano.index)
axs['A'].set_xticklabels(labels = df_agrupado_ano.index, fontweight = 'bold', fontsize = 12)
axs['A'].tick_params(axis = 'x', length = 0)
axs['A'].set_yticks([])

axs['A'].set_title('Quantidade de homicídios por ano no Brasil\n', color = 'Darkgray', fontsize = 16, fontweight = 'bold')


axs['A'].set_frame_on(False)

axs['A'].grid(True, color='gray', linestyle='--', linewidth= 1, alpha= 0.2)


# gráfico 2 - 

barras = axs['B'].barh(estados_ordenados, mortes_ordenadas, color = 'bisque', alpha = 0.4)

rotulos =  [f"{morte:,.0f}" for morte in mortes_ordenadas]

for i , rotulo in enumerate(rotulos):
    if i > 1:
        rotulo = ''
        rotulos[i] = rotulo
    else: pass
        
axs['B'].bar_label(barras, labels= rotulos, label_type='edge', padding= 1, fontweight = 'bold')

axs['B'].set_xticks([])
axs['B'].tick_params(axis = 'y', length = 0)
axs['B'].set_yticklabels(labels = estados_ordenados,fontweight = 'bold', fontsize = 12 )
# axs['B'].set_yticks(fontweight = 'bold', fontsize = 12)

barras[0].set_color(cores[4])
barras[0].set_alpha(1)

barras[1].set_color(cores[4])
barras[1].set_alpha(1)
        

axs['B'].set_title('Estados com maiores índices de homicídio ', color = 'Darkgray', fontsize = 16, fontweight = 'bold')


axs['B'].invert_yaxis()

axs['B'].set_frame_on(False)

# ----------------------------------------

# gráfico 3 - 

df_agrupado_menores_ind = df_final.groupby('Estado')['Mortes totais'].sum().sort_values(ascending = True).to_frame()

df_agrupado_menores_ind = df_agrupado_menores_ind.head()

dados_ordenados = sorted(zip(df_agrupado_menores_ind['Mortes totais'], df_agrupado_menores_ind.index), reverse = True)
mortes_ordenadas = [x[0] for x in dados_ordenados]
estados_ordenados = [x[1] for x in dados_ordenados]

df_agrupado_menores_ind

barras = axs['C'].barh(estados_ordenados, mortes_ordenadas, color = 'bisque', alpha = 0.4)

rotulos =  [f"{morte:,.0f}" for morte in mortes_ordenadas]

for i , rotulo in enumerate(rotulos):
    if i < 3:
        rotulo = ''
        rotulos[i] = rotulo
    else: pass
        
axs['C'].bar_label(barras, labels= rotulos, label_type='edge', padding= 1, fontweight = 'bold')

axs['C'].set_xticks([])
axs['C'].tick_params(axis = 'y', length = 0)
axs['C'].set_yticklabels(labels = estados_ordenados, fontweight = 'bold', fontsize = 12)

barras[-2].set_color(cores[0])
barras[-2].set_alpha(1)

barras[-1].set_color(cores[0])
barras[-1].set_alpha(1)
        

axs['C'].set_title('Estados com menores índices de homicídio ', color = 'Darkgray', fontsize = 16, fontweight = 'bold')

axs['C'].set_frame_on(False)


# gráfico 4 -

df_mulheres.rename(columns= {'sigla_uf_nome' : 'Estado'}, inplace= True)

df_agrupado_mulheres = df_mulheres.groupby('Estado')[['quantidade_estupro', 'quantidade_feminicidio']].sum()

df_agrupado_mulheres

axs['D'].scatter('quantidade_estupro', 'quantidade_feminicidio', data = df_agrupado_mulheres, s = 'quantidade_feminicidio', c =  'quantidade_feminicidio',cmap = "Reds" )
# axs['D'].scatter(ano, receita, s = receita, c = receita, cmap = "Reds")

axs['D'].grid(True, linestyle = "--")
axs['D'].set_ylabel("Feminicídios", fontweight = "bold", color = 'Darkgray', fontsize = 16)
axs['D'].set_xlabel("Estrupos", fontweight = "bold", color = 'Darkgray', fontsize = 16)
axs['D'].set_title("Quantidade de Estrupos x Quantidade de Feminicídio\n\nAgrupado por Estado\n", fontweight = "bold", color = 'Darkgray', fontsize = 16)


mappable = axs['D'].scatter('quantidade_estupro', 'quantidade_feminicidio', data = df_agrupado_mulheres, c='quantidade_feminicidio', cmap="Reds")
fig.colorbar(mappable, ax = axs['D'])

axs['D'].set_frame_on(False)


fig.suptitle('Anuário Brasileiro de Segurança Pública', fontsize = 20, color = 'black', fontweight = 'bold')

fig.savefig('Relatório segurança feito.png', bbox_inches = 'tight', dpi = 300)

plt.show()
```
